```python
import pandas as pd

df = pd.read_csv('../paper2/total/true_detections_full.csv')

# Remove rows where Noise contains non-numeric text (e.g. 'Î¼Jy/beam')
df_clean = df[df['Noise'].str.contains(r'^\d+(\.\d+)?$', regex=True, na=False)]

# Convert to numeric
df_clean['Noise'] = pd.to_numeric(df_clean['Noise'])

# Compute median
median_noise = df_clean['Noise'].median()

print("Median Noise =", median_noise)
```
