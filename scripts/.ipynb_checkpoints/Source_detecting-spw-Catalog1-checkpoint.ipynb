```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
frequencies = [1.6535753, 1.7175753, 1.7815753, 1.8455753, 1.9095753, 2.0376753, 0.9615883, 1.3455883, 1.4095883]
for i in spw:
    df['SPW'] = i
    df['Frequency'] = frequencies[spw.index(i)]
    df['Total_flux'] = df['Total_flux'] * 1e3  # Convert flux to mJy
    dataframes.append(df)
all_sources = pd.concat(dataframes)
grouped_sources = all_sources.groupby(['RA_rounded', 'DEC_rounded'])
def power_law(x, a, b):
    return a * x ** b
def calculate_spectral_index(group):
    if len(group) < 2:
        return np.nan  # Cannot calculate spectral index with less than 2 measurements
    freqs = group['Frequency'].values
    fluxes = group['Total_flux'].values
    try:
        popt, _ = curve_fit(power_law, freqs, fluxes, maxfev=10000)
        return popt[1]  # The spectral index is the exponent 'b' in the power-law
    except RuntimeError:
        return np.nan
filtered_groups = grouped_sources.filter(lambda x: x['SPW'].nunique() > 6)
spectral_indices = filtered_groups.groupby(['RA_rounded', 'DEC_rounded']).apply(calculate_spectral_index).reset_index()
spectral_indices.columns = ['RA_rounded', 'DEC_rounded', 'Spectral_Index']
for i in range(len(dataframes)):
    dataframes[i].to_csv(f'../paper1/csv/spw{spw[i]}.csv', index=False)
print("Spectral index calculation completed for sources appearing in more than 6 SPWs and updated dataframes saved to CSV files.")
```

# Comparing two methods

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
def power_law(x, a, b):
    return a * x ** b
def calculate_spectral_index_polyfit(group):
    if len(group) < 2:
        return np.nan  # Cannot calculate spectral index with less than 2 measurements
    log_freqs = np.log(freqs)
    log_fluxes = np.log(fluxes)
    slope, _ = np.polyfit(log_freqs, log_fluxes, 1)
    return slope
def calculate_spectral_index_powerlaw(group):
    if len(group) < 2:
        return np.nan  # Cannot calculate spectral index with less than 2 measurements
    try:
        return popt[1]  # Return the exponent as the spectral index
    except RuntimeError:
        return np.nan
spectral_indices_polyfit = filtered_groups.groupby(['RA_rounded', 'DEC_rounded']).apply(calculate_spectral_index_polyfit).reset_index()
spectral_indices_polyfit.columns = ['RA_rounded', 'DEC_rounded', 'Spectral_Index_Polyfit']
spectral_indices_powerlaw = filtered_groups.groupby(['RA_rounded', 'DEC_rounded']).apply(calculate_spectral_index_powerlaw).reset_index()
spectral_indices_powerlaw.columns = ['RA_rounded', 'DEC_rounded', 'Spectral_Index_Powerlaw']
spectral_indices_combined = pd.merge(spectral_indices_polyfit, spectral_indices_powerlaw, on=['RA_rounded', 'DEC_rounded'])
```

```python
def calculate_rss(freqs, fluxes, fit_line):
    residuals = fluxes - fit_line
    rss = np.sum(residuals**2)
    return rss
def calculate_rss_polyfit(group):
    if len(group) < 2:
        return np.nan
    slope, intercept = np.polyfit(log_freqs, log_fluxes, 1)
    return calculate_rss(freqs, fluxes, fit_line)
def calculate_rss_powerlaw(group):
    if len(group) < 2:
        return np.nan
    try:
        return calculate_rss(freqs, fluxes, fit_line)
    except RuntimeError:
        return np.nan
rss_polyfit = filtered_groups.groupby(['RA_rounded', 'DEC_rounded']).apply(calculate_rss_polyfit).reset_index()
rss_polyfit.columns = ['RA_rounded', 'DEC_rounded', 'RSS_Polyfit']
rss_powerlaw = filtered_groups.groupby(['RA_rounded', 'DEC_rounded']).apply(calculate_rss_powerlaw).reset_index()
rss_powerlaw.columns = ['RA_rounded', 'DEC_rounded', 'RSS_Powerlaw']
spectral_indices_combined['Better_Method'] = np.where(
    spectral_indices_combined['RSS_Polyfit'] < spectral_indices_combined['RSS_Powerlaw'],
    'Polyfit',
    'Powerlaw'
)
print(spectral_indices_combined.head())
```

```python
spectral_indices_combined.to_csv('../paper1/csv/Spectral_Indices_Comparison.csv', index=False)
print("Spectral indices calculated using both methods and compared. Results saved to CSV file.")
```

```python
total_sources = len(spectral_indices_combined)
better_powerlaw_count = np.sum(spectral_indices_combined['Better_Method'] == 'Powerlaw')
percentage_better_powerlaw = (better_powerlaw_count / total_sources) * 100
print(f"Percentage of sources where the power-law method is better: {percentage_better_powerlaw:.2f}%")
```

# Latex

```python
updated_dataframes = []
for s in spw:
    updated_dataframes.append(df)
    print(f"Shape of DataFrame for SPW {s} after reading from CSV: {df.shape}")
print("Reading process completed.")
```

```python
loc_columns = ['RA', 'E_RA', 'DEC', 'E_DEC', 'Spectral_Index']
flux_columns = ['Total_flux', 'E_Total_flux', 'Peak_flux', 'E_Peak_flux', 'Isl_rms']
```

```python
processed_dataframes = []
for df in updated_dataframes:
    selected_columns = df[loc_columns + flux_columns].copy()
    selected_columns[flux_columns] = selected_columns[flux_columns] * 1e3
    selected_columns['Spectral_Index'] = selected_columns['Spectral_Index'].map('{:.2f}'.format)
    selected_columns[['RA', 'E_RA', 'DEC', 'E_DEC']] = selected_columns[['RA', 'E_RA', 'DEC', 'E_DEC']].applymap(lambda x: f"{x:.5f}")
    processed_dataframes.append(selected_columns)
for i, df in enumerate(processed_dataframes):
    print(f"Processed DataFrame for SPW {spw[i]}:")
    print(df.head())
for i, df in enumerate(processed_dataframes):
    df.to_csv(f'../paper1/csv/processed_spw{spw[i]}.csv', index=False)
print("Processing and saving completed.")
```

```python
rename_dict = {
    'RA': r'\text{RA}',
    'E_RA': r'$\sigma_{\text{RA}}$',
    'DEC': r'\text{DEC}',
    'E_DEC': r'$\sigma_{\text{DEC}}$',
    'Total_flux': r'$S_{\text{total}}$',
    'E_Total_flux': r'$\sigma_{S_{\text{total}}}$',
    'Peak_flux': r'$S_{\text{peak}}$',
    'E_Peak_flux': r'$\sigma_{S_{\text{peak}}}$',
    'Isl_rms':r'$RMS$',
    'Spectral_Index':r'$\alpha$'
}
```

```python
rows_per_chunk=60
for i, df in enumerate(processed_dataframes):
    df.rename(columns=rename_dict, inplace=True)
    num_chunks = (len(df) // rows_per_chunk) + 1
    for chunk_index in range(num_chunks):
        start_row = chunk_index * rows_per_chunk
        end_row = start_row + rows_per_chunk
        chunk_df = df.iloc[start_row:end_row]
        latex_table = chunk_df.style.hide(axis='index').to_latex(hrules=True)
        with open(f'../paper1/csv/latex/spw{spw[i]}_part{chunk_index + 1}.tex', 'w') as file:
            file.write(latex_table)
print("Processing and saving to LaTeX longtables completed.")
```

```python
import pandas as pd
import re
def parse_entry(first_line, second_line):
    primary_pattern = re.compile(
        r'(\d{2})\s+(\d{2})\s+([\d\.]+)\s+([+-]?\d{2})\s+(\d{2})\s+([\d\.]+)\s+([\d\.]+)\s+([\d<>\.]+)\s+([\d<>\.]+)\s*([\d\.\*]*)\s*([\d\.\*]*)\s*([\d\.\-]*)\s*([\d\.\-]*)\s*([a-zA-Z0-9\*]*)\s*([\d\.\-]*)\s*([\d\.\-]*)'
    )
    error_pattern = re.compile(
        r'\s*([\d\.]*)\s+([\d\.]*)\s+([\d\.]*)\s+([\d\.]*)\s+([\d\.]*)\s*([\d\.\-]*)\s*([\d\.\-]*)\s*([\d\.\-]*)\s*([\d\.\-]*)'
    )
    primary_match = primary_pattern.match(first_line)
    error_match = error_pattern.match(second_line)
    if primary_match and error_match:
        return {
            'RA_h': primary_match.group(1),
            'RA_m': primary_match.group(2),
            'RA_s': primary_match.group(3),
            'RA_s_err': error_match.group(1),
            'Dec_d': primary_match.group(4),
            'Dec_m': primary_match.group(5),
            'Dec_s': primary_match.group(6),
            'Dec_s_err': error_match.group(2),
            'Flux (mJy)': primary_match.group(7),
            'Flux_err (mJy)': error_match.group(3),
            'Major (")': primary_match.group(8),
            'Major_err (")': error_match.group(4),
            'Minor (")': primary_match.group(9),
            'Minor_err (")': error_match.group(5),
            'PA (deg)': primary_match.group(10),
            'PA_err (deg)': error_match.group(6),
            'Res': primary_match.group(11),
            'Res_err': error_match.group(7),
            'P_Flux (mJy)': primary_match.group(12),
            'P_Flux_err (mJy)': error_match.group(8),
            'P_ang (deg)': primary_match.group(13),
            'P_ang_err (deg)': error_match.group(9),
            'Field': primary_match.group(14),
            'X_pix': primary_match.group(15),
            'Y_pix': primary_match.group(16)
        }
    else:
        return None
file_path = '../paper1/NVSSCatalog.text'
with open(file_path, 'r') as file:
    lines = file.readlines()
filtered_lines = []
for line in lines:
    if re.match(r'^\s*\d{2}\s+\d{2}\s+\d', line) or re.match(r'^\s*\d+\.\d+', line):
        filtered_lines.append(line)
if len(filtered_lines) % 2 != 0:
    filtered_lines.append(' ' * len(filtered_lines[0]))
entries = []
for i in range(0, len(filtered_lines), 2):
    if i + 1 < len(filtered_lines):
        entries.append((filtered_lines[i], filtered_lines[i + 1]))
data = []
for entry in entries:
    parsed_entry = parse_entry(entry[0], entry[1])
    if parsed_entry:
        data.append(parsed_entry)
problematic_rows = df[df['Dec_s'].str.contains(r'[^0-9\.]', na=False)]
print("Problematic rows:")
print(problematic_rows)
def clean_dec_s(value):
    parts = value.split('.')
    if len(parts) > 2:
        return parts[0] + '.' + parts[1]  # Keep the first two parts
    return value
df['Dec_s'] = df['Dec_s'].apply(clean_dec_s)
df['Dec_s_err'] = pd.to_numeric(df['Dec_s_err'], errors='coerce')
df.to_csv('../paper1/csv/NVSS.csv', index=False)
```

## comparing


```python
df_nvss = pd.read_csv('../paper1/csv/NVSS_with_degrees.csv')
df_spw16 = pd.read_csv('../paper1/csv/spw16.csv')
nvss_ra = df_nvss['RA_deg'].values
nvss_dec = df_nvss['Dec_deg'].values
nvss_ra_err = df_nvss['RA_err_deg'].values
nvss_dec_err = df_nvss['Dec_err_deg'].values
spw16_ra = df_spw16['RA'].values
spw16_dec = df_spw16['DEC'].values
spw16_ra_err = df_spw16['E_RA'].values
spw16_dec_err = df_spw16['E_DEC'].values
matches = []
for i in range(len(nvss_ra)):
    ra_diff = np.abs(nvss_ra[i] - spw16_ra)
    dec_diff = np.abs(nvss_dec[i] - spw16_dec)
    ra_err_sum = nvss_ra_err[i] + spw16_ra_err
    dec_err_sum = nvss_dec_err[i] + spw16_dec_err
    ra_match = ra_diff <= ra_err_sum
    dec_match = dec_diff <= dec_err_sum
    match_indices = np.where(ra_match & dec_match)[0]
    for j in match_indices:
        matches.append({
            'NVSS_index': i,
            'spw16_index': j,
            'NVSS_RA_deg': nvss_ra[i],
            'NVSS_Dec_deg': nvss_dec[i],
            'spw16_RA_deg': spw16_ra[j],
            'spw16_Dec_deg': spw16_dec[j]
        })
df_matches = pd.DataFrame(matches)
df_matches.to_csv('../paper1/csv/NVSS_spw16_matches.csv', index=False)
print(df_matches.head())
```

```python
df_matches
```

```python
import pandas as pd
matched_sources = []
for _, match in df_matches.iterrows():
    nvss_index = int(match['NVSS_index'])
    spw16_index = int(match['spw16_index'])
    nvss_row = df_nvss.iloc[nvss_index]
    spw16_row = df_spw16.iloc[spw16_index]
    matched_sources.append({
        'NVSS_index': nvss_index,
        'spw16_index': spw16_index,
        'NVSS_RA_deg': nvss_row['RA_deg'],
        'NVSS_Dec_deg': nvss_row['Dec_deg'],
        'spw16_RA_deg': spw16_row['RA'],
        'spw16_Dec_deg': spw16_row['DEC'],
        'NVSS_Flux (mJy)': nvss_row['Flux (mJy)'],
        'NVSS_Flux_err (mJy)': nvss_row['Flux_err (mJy)'],
        'spw16_Flux (mJy)': spw16_row['Total_flux'] * 1e3,  # Convert to mJy
        'spw16_Flux_err (mJy)': spw16_row['E_Total_flux'] * 1e3  # Convert to mJy
    })
df_matched_sources = pd.DataFrame(matched_sources)
df_matched_sources.to_csv('../paper1/csv/NVSS_spw16_fluxes.csv', index=False)
print(df_matched_sources.head())
```

```python
import pandas as pd
import numpy as np
def flux_match(nvss_flux, nvss_flux_err, spw16_flux, spw16_flux_err):
    return np.abs(nvss_flux - spw16_flux) <= (nvss_flux_err + spw16_flux_err)
df_matched_sources['Fluxes_Match'] = df_matched_sources.apply(
    lambda row: 'Yes' if flux_match(
        float(row['NVSS_Flux (mJy)']), float(row['NVSS_Flux_err (mJy)']),
        float(row['spw16_Flux (mJy)']), float(row['spw16_Flux_err (mJy)'])
    ) else 'No', axis=1
)
df_matched_sources.to_csv('../paper1/csv/NVSS_spw16_flux_comparison.csv', index=False)
print(df_matched_sources.head())
```

```python
import pandas as pd
def calculate_percentage_difference(nvss_flux, spw16_flux):
    try:
        return abs(nvss_flux - spw16_flux) / nvss_flux * 100
    except ZeroDivisionError:
        return float('inf')  # Handle the case where NVSS flux is zero
df_no_matches['Percentage_Difference'] = df_no_matches.apply(
    lambda row: calculate_percentage_difference(row['NVSS_Flux (mJy)'], row['spw16_Flux (mJy)']), axis=1
)
df_no_matches.to_csv('../paper1/csv/NVSS_spw16_no_matches_percentage_difference.csv', index=False)
print(df_no_matches[['NVSS_index', 'spw16_index', 'NVSS_Flux (mJy)', 'spw16_Flux (mJy)', 'Percentage_Difference']])
```
