```python
import numpy as np
import matplotlib.pyplot as plt

# Use a dark background style
plt.style.use('dark_background')

# Generate x values
x_values = np.array([90, 500, 800, 1100])

# Actual y values
y_values = np.array([140, 150, 155, 158]) 
y_errors = np.array([0.5, 0.3, -2, 0.2])  # small error to simulate measurement errors
y_values_with_error = y_values + y_errors

# Define small error bars
y_err = [2] * len(x_values)  # consistent small error bars

# Fit a linear model to the data
coefficients = np.polyfit(x_values, y_values, 1)
polynomial = np.poly1d(coefficients)

# Extract slope and intercept
slope, intercept = coefficients
print("Slope of the line: {:.2f}".format(slope))
print("Intercept of the line: {:.2f}".format(intercept))

# Generate y values for the linear fit line over the range of x_values
x_fit = np.linspace(min(x_values), max(x_values), 100)
y_fit = polynomial(x_fit)

# Plotting
plt.figure(figsize=(8, 5))
plt.errorbar(x_values, y_values_with_error, yerr=y_err, fmt='o', capsize=5, color='white', label='Data with error bars')
plt.plot(x_fit, y_fit, label='Fitted linear line', color='red', linestyle='--')  # Dashed line style

plt.title('$RM [rad/m^{-2}$] is the slope of the line', fontsize='15')
plt.xlabel(r'$\lambda^{2}$', fontsize='15')
plt.ylabel(r'$\Psi(degrees)$', fontsize='15')
plt.legend()

# Set custom x-axis ticks at the locations where your data exists
plt.xticks(x_values, ['A', 'B', 'C', 'D'])
plt.savefig('RM plot')

plt.show()
```
