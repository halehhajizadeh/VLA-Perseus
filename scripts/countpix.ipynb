```python
from astropy.io import fits
import numpy as np
```

```python
fitsname = '../server/total/combined_result.fits'
```

```python
# Open the FITS file and get the data
with fits.open(fitsname) as hdul:
    data = hdul[0].data


total_pixels = data.size


non_nan_pixels = np.count_nonzero(~np.isnan(data))


percentage_non_nan = (non_nan_pixels / total_pixels) * 100

# print(f"Total number of pixels: {total_pixels}")
print(f"Number of non-NaN pixels: {non_nan_pixels}")
# print(f"Percentage of non-NaN pixels: {percentage_non_nan:.2f}%")
```
